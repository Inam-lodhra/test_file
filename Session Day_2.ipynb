```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set theme for better visualization
sns.set_theme(style="ticks", color_codes=True)

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a count plot
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
p1 = sns.countplot(x="sex", data=titanic, hue="class")

# Set plot title
p1.set_title("Passenger Count by Sex and Class")

# Show the plot
plt.show()
```

```python
# scatter plot
import seaborn as sns
import matplotlib.pyplot as plt

# Set theme for better visualization
sns.set_theme(style="ticks", color_codes=True)

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

g=sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age", "fare").add_legend())
plt.show()
```
