```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ship= sns.load_dataset("titanic")
sns.boxplot(x= "survived",
            y= "age", 
            showmeans=True,
            meanprops= {"marker": "*",
                       "markersize": "12",
                       "markeredgecolor": "red"},
            data= ship)

plt.show()
```
