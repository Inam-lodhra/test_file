# Horizontal bar plot

```python
# import libraries
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

# load dataset
ship= sns.load_dataset("titanic")
ship

# draw a line plot
sns.barplot(x= "fare", y= "class", hue= "sex", data= ship, estimator=median, saturation=1.5) # ci is now changed to errorbar
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
ship = sns.load_dataset("titanic")

# Create bar plot with corrected arguments
sns.barplot(x="class", y="fare", data=ship,
            linewidth=1, facecolor=(0.1, 0.3, 0., 1),
            errcolor=".5", edgecolor=".1")

# Show plot
plt.show()
```
